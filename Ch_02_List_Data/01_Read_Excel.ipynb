```python
import matplotlib.pyplot as plt
import numpy as np
data = np.loadtxt("/Users/sandeepdash/Desktop/Calories.txt", delimiter=",")
ypoints = np.array(data)

# for full year (days 1 - 365)
plt.plot(ypoints, linestyle = 'dotted')
plt.show()

# for Q3 (days 183 - 274)
indices = list()
for n in range(183,274):
    indices.append(n)

new_array = np.take(ypoints, indices)
plt.plot(new_array, linestyle = 'solid')
plt.show()
```
