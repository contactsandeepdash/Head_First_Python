```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("/Users/sandeepdash/Desktop/test.csv", delimiter=",")
ypoints = np.array(data)

print("Index of min value = " + str(min(ypoints)) + " is: ",np.where(ypoints == min(ypoints))[0][0])
print("Index of max value = " + str(max(ypoints)) + " is: ",np.where(ypoints == max(ypoints))[0][0])

# for full year (days 1 - 365)
#plt.plot(ypoints, linestyle = 'dotted')
#plt.show()

# for Q3 (days 181 - 273)
indices = list()
for n in range(181,273):
    indices.append(n)

new_array = np.take(ypoints, indices)
#plt.plot(indices, new_array, linestyle = 'solid', color="red")
#plt.show()

def annot_max(y, ax=None):
    ymax = y.max()
    text= "y={:.3f}".format(ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, y=(ymax), xytext=(0.94,0.96), **kw)

annot_max(ypoints)

fig, ax = plt.subplots()
ax.plot(indices, new_array, linestyle = 'solid', color="red")
plt.figure(figsize=(10,6))
ax.set_ylim(-0.3,1.5)
plt.show()


#plt.figure()
#plt.title('Horizontal Box Plot')
#flierprops = dict(marker='o', markerfacecolor='r', markersize=12,
#                  linestyle='none', markeredgecolor='g')
#plt.boxplot(new_array, showmeans=True, meanline=True, notch=False, vert=False, flierprops=flierprops)
#plt.ylabel('Variation')
#plt.show()


```
